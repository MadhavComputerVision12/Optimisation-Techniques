```python
from pyswarm import pso
import numpy as np

# Define the objective function based on the polynomial equation
def objective_function(x):
    return abs(2*x[0]**2 - 5*x[0] + 2)

# Set up the search space (based on expected range of roots)
lb = [0]  # Lower bound for the root
ub = [5]  # Upper bound for the root

# Run PSO to minimize the objective function
x_opt, f_opt = pso(objective_function, lb, ub, swarmsize=10, maxiter=100)

print("Optimal root:", x_opt[0])
```
